```python
from pyspark.sql import SparkSession

# Initialize Spark session
spark = SparkSession.builder.appName('WordCount').getOrCreate()
```

```python
# Load text file
text_rdd = spark.sparkContext.textFile('Sample-text-file.txt')
```

```python
# Split each line into words
words_rdd = text_rdd.flatMap(lambda line: line.split())
```

```python
# Map each word to a tuple (word, 1)
word_tuples = words_rdd.map(lambda word: (word, 1))

# Reduce by key to count occurrences
word_counts = word_tuples.reduceByKey(lambda a, b: a + b)
```

```python
# Collect the results
word_counts_collected = word_counts.collect()

# Display the results
for word, count in word_counts_collected:
    print(f'{word}: {count}')

# Stop Spark session
spark.stop()
```
